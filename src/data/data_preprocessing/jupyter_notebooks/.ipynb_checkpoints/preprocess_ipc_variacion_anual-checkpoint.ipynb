# Imports

```python
import pandas as pd
import numpy as np
```

```python
ipc_variacion_anual_df = pd.read_excel(io = './../original_data/bde_banco_central/ipc_variacion_anual.xlsx', skiprows = 2, 
                                       names = ['Periodo', 'IPC General'])
ipc_variacion_anual_df
```

# Data Analysis

```python
ipc_variacion_anual_df.info()
```

```python
ipc_variacion_anual_df[ipc_variacion_anual_df['IPC General'].isna()]
```

```python
ipc_variacion_anual_df = ipc_variacion_anual_df[ipc_variacion_anual_df['Periodo'] > pd.Timestamp(1929,2,1)]
ipc_variacion_anual_df
```

```python
ipc_variacion_anual_df = ipc_variacion_anual_df[ipc_variacion_anual_df['Periodo'].dt.month == 6]
ipc_variacion_anual_df
```

# Save CSV file

```python
# Set the date format as the original data (%Y-%m), without day
ipc_variacion_anual_df['Periodo'] = ipc_variacion_anual_df['Periodo'].dt.to_period('Y')
ipc_variacion_anual_df
```

```python
ipc_variacion_anual_df.to_csv('./../preprocessed_data/ipc_variacion_anual_junio.csv', sep = ',', index = False)
```
